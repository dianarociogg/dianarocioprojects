# Seaborn: Datos Categóricos

A continuación veremos visualizaciones especiales para datos categóricos:

* barplot
* countplot
* boxplot
* violinplot
* stripplot
* swarmplot
* factorplot


```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## barplot y countplot

Estas gráficas son muy similares, permiten obtener datos agregados de una variable categórica. 

### barplot

** barplot ** es un gráfico general que le permite agregar los datos categóricos basados en alguna función, por defecto la media:

```python
sns.barplot(x='sex',y='total_bill',data=tips) # ci = sd (cambiaría el propósito de la línea), así como está es el error promedio
# de la medida, intervalo de confianza, ci=None sería sin línea
```

```python
sns.barplot(x='sex',y='tip',data=tips)
```

Es posible cambiar la función de agregación, por una función propia u otra predefinida que convierta un vector a un escalar, por ejemplo, la desviación estándar:

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std) # Los datos que presenta en las columnas no es el promedio si no 
# la desviación estandar 
```

### countplot

Básicamente es lo mismo que barplot, excepto que el estimador está contando explícitamente el número de ocurrencias. Por eso solo pasamos el valor de x:

```python
sns.countplot(x='sex',data=tips) # contar cuantos elementos hay por categoría
```

## boxplot y violinplot

Los diagramas de caja (boxplots) y los "violinplots" se utilizan para mostrar la distribución de datos categóricos. 


### boxplot 

Un diagrama de caja muestra la distribución de datos cuantitativos de una manera que facilita las comparaciones entre variables o entre niveles de una variable categórica. El cuadro muestra los cuartiles del conjunto de datos, mientras que los bigotes se extienden para mostrar el resto de la distribución, excepto los puntos que se determinan como valores atípicos (outliers) utilizando un método que es una función del rango intercuartílico.


```python
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
# Pueden representar el DataFrame completo con orient='h'
sns.boxplot(data=tips,palette='rainbow',orient='h')
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm")
```

### violinplot
Un gráfico de violín juega un papel similar al diagrama de cajas. Muestra la distribución de datos cuantitativos a través de varios niveles de una (o más) variables categóricas de modo que esas distribuciones se puedan comparar. A diferencia de un diagrama de caja, en el que todos los componentes de la gráfica corresponden a puntos de datos reales, la gráfica del violín presenta una estimación de la densidad del núcleo de la distribución subyacente.

```python
sns.violinplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',palette='Set1')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',split=True,palette='Set1') # Pone a lado y lado la distribución 
# de las categorías no en figuras diferentes
```

## stripplot y swarmplot

### stripplot
El stripplot dibuja un diagrama de dispersión donde una variable es categórica. Se puede dibujar por sí mismo, pero también es un buen complemento de un diagrama de caja o de violín en los casos en que deseen mostrar todas las observaciones junto con alguna representación de la distribución subyacente.

```python
sns.stripplot(x="day", y="total_bill", data=tips) 
#sns.stripplot(x="day", y="tip", data=tips)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True) # pinta el mismo stripplot solo que pone la línea más ancha para 
# ver mejor los puntos
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1') # Ponga más ancha la representación y el 
# tono pongalo de acuerdo a la variable categórica sexo
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1',dodge=True) 
# el parámetro dodge antes se llamaba split, no mezclar los puntos de tonos distintos
```

### swarmplot

El swarmplot (enjambre) es similar a stripplot(), pero los puntos se ajustan (solo a lo largo del eje categórico) para que no se superpongan. Esto proporciona una mejor representación de la distribución de los valores, aunque no se ajusta a un gran número de observaciones (tanto en términos de la capacidad de mostrar todos los puntos como en términos del cálculo necesario para organizarlos).

```python
sns.swarmplot(x="day", y="total_bill", data=tips) # grafica todos los puntos, en el anterior se podían sobreponer, funciona solo
# en casos donde son pocos datos
```

```python
sns.swarmplot(x="day", y="total_bill",hue='sex',data=tips, palette="Set1", dodge=True)
```

## Combinaciones de gráficos para variables categóricas

```python
sns.violinplot(x="tip", y="day", data=tips,palette='rainbow') # al poner la variable numérica en el eje x, el violín se despliega
# de forma vertical
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=3)
```

## factorplot

factorplot es el método más general para visualizar datos categóricos. Recibe el parámetro **kind** para seleccionar el tipo de gráfico a presentar:

```python
sns.factorplot(x='sex',y='total_bill',data=tips,kind='bar') # generalización de todos los anteriores
```

```python
sns.factorplot(x='sex',y='total_bill',data=tips,kind='swarm')
```
