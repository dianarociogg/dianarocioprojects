# Gráficos de distribuciones

A continuación se presentan algunos tipos de gráficos de Seaborn útiles para visualizar la distribución de los datos. Estos son:

* distplot
* jointplot
* pairplot
* rugplot
* kdeplot

___
## Importar

Como está construida sobre matplotlib también requiere el comando "inline"

```python
import seaborn as sns
%matplotlib inline
```

## Datos
Seaborn posee algunos datasets. En este caso trabajaremos con el de propinas (tips).

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## distplot

Distplot muestra la distribución de un conjunto de observaciones de una variable.

```python
sns.distplot(tips['total_bill'])
# Ignorar los warnings
```

Por defecto presenta la figura de Estimación de Densidad de Kernel (KDE). Para quitarla simplemente añada el siguiente parámetro: `kde=False`

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

## jointplot

jointplot() le permite básicamente emparejar dos distplots para datos bivariados. Con su elección del parámetro **kind** permite comparar diferentes visualizaciones: 
* “scatter” 
* “reg” 
* “resid” 
* “kde” 
* “hex”

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter') 
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex') # Más util cuando se superponen datos, para 
# saber donde caen la mayoría de mis datos
# buscar como cambiar el tamaño del hexagono!
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde') # combina las dos estimaciones de la 
# densidad de kernel 
```

## pairplot

pairplot grafica las distribuciones comparando relaciones de las variables del DataFrame, presentando todas las combinaciones de 2 variables (para las columnas numéricas). Acepta un argumento de tono de color - `hue` (para columnas categóricas).


```python
tips.head()
```

```python
sns.pairplot(tips) # hace una cuadrícula de figuras para las variables numéricas del DF
```

```python
sns.pairplot(tips,hue='sex') # Se pueden filtrar datos usando solo el tono :)
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm') # mapa de color por defecto de matplotlib 
```

## rugplot

Los rugplots son una representación muy simple. Sólo dibujan una marca para cada punto en una distribución univariada. Ellos son el componente básico de una gráfico KDE (Kernel Density Distribution):

```python
sns.rugplot(tips['total_bill'])
```

## kdeplot

kdeplots son [gráficos de Estimación de Densidad del Kernel](http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth). Estas representaciones de KDE reemplazan cada observación con una distribución Gaussiana (Normal) centrada alrededor de ese valor. Por ejemplo:


```python
# NO ES NECESARIO ENTEDER ESTE CODIGO
# SOLO SE UTILIZARÁ PARA EXPLICAR KDE
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# dataset
dataset = np.random.randn(25)

# Create another rugplot
sns.rugplot(dataset);

# Set up the x-axis for the plot
x_min = dataset.min() - 2
x_max = dataset.max() + 2

# 100 equally spaced points from x_min to x_max
x_axis = np.linspace(x_min,x_max,100)

# Set up the bandwidth, for info on this:
url = 'http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth'

bandwidth = ((4*dataset.std()**5)/(3*len(dataset)))**.2


# Create an empty kernel list
kernel_list = []

# Plot each basis function
for data_point in dataset:
    
    # Create a kernel for each point and append to list
    kernel = stats.norm(data_point,bandwidth).pdf(x_axis)
    kernel_list.append(kernel)
    
    #Scale for plotting
    kernel = kernel / kernel.max()
    kernel = kernel * .4
    plt.plot(x_axis,kernel,color = 'grey',alpha=0.5)

plt.ylim(0,1)
```

```python
# Para obtener el KDE se suman las funciones base

# Plot the sum of the basis function
sum_of_kde = np.sum(kernel_list,axis=0)

# Plot figure
fig = plt.plot(x_axis,sum_of_kde,color='indianred')

# Add the initial rugplot
sns.rugplot(dataset,c = 'indianred')

# Get rid of y-tick marks
plt.yticks([])

# Set title
plt.suptitle("Sum of the Basis Functions")
```

Entonces, usando nuestro dataset (tips):
