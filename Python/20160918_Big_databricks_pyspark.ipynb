```
#Recategorizar los rangos de edades
from pyspark.sql.types import IntegerType
from pyspark.sql.functions import when, lit, col

def set_range(RANGOEDAD):
  if (RANGOEDAD != RANGOEDAD) | (RANGOEDAD is None):
    return None
  
  if RANGOEDAD == 'Entre 21 y 25 a�os':
    return 1
  if RANGOEDAD == 'Entre 26 y 30 a�os':
    return 1  

  if RANGOEDAD == 'Entre 31 y 35 a�os':
    return 2
  if RANGOEDAD == 'Entre 36 y 40 a�os':
    return 2  

  if RANGOEDAD == 'Entre 41 y 45 a�os':
    return 3
  if RANGOEDAD == 'Entre 46 y 50 a�os':
    return 3

  if RANGOEDAD == 'Entre 51 y 55 a�os':
    return 4
  if RANGOEDAD == 'Entre 56 y 60 a�os':
    return 4
  
  if RANGOEDAD == 'Entre 61 y 65 a�os':
    return 5
  if RANGOEDAD == 'Entre 66 y 70 a�os':
    return 5
  
  if RANGOEDAD == 'Entre 71 y 75 a�os':
    return 6
  if RANGOEDAD == 'Mayor a 75 a�os':
    return 6
    
udfRange = udf(set_range, IntegerType())
customers = customers.withColumn('RANGOEDADREC', udfRange('RANGOEDAD'))
display(customers)
```

```
#Ver los distintos rangos de edades recategorizados de los clientes en el conjunto de datos:
display(customers.select('RANGOEDADREC').groupBy('RANGOEDADREC').count().orderBy("count",ascending=False))
```

#### 1.¿En qué rango de edad se pide un crédito para compra de vivienda?

```
#1.¿En qué rango de edad se pide un crédito para compra de vivienda?
edadCVDF = customers.where((col("DESTINOECONOMICO") == "COMPRA DE VIV. CREDITO CONSTRUCTOR") | (col("DESTINOECONOMICO") == "COMPRA VIV. LIBERACION FNA") | (col("DESTINOECONOMICO") == "COMPRA VIV. LIBERACION FNA CORPORACION") | (col("DESTINOECONOMICO") == "COMPRA VIVIENDA") | (col("DESTINOECONOMICO") == "COMPRA VIVIENDA CON LIBERACION") | (col("DESTINOECONOMICO") == "COMPRA VIVIENDA VIS")
)
#print(edadCVDF.count())--67003
display(edadCVDF.select('RANGOEDADREC').groupBy('RANGOEDADREC').count().orderBy("count",ascending=False))
```

El mayor número de solicitud de crédito se presentó en personas en Entre 41 y 50 años

#### 4. ¿Para qué se solicitan más créditos en cada Departamento?

```
## Identificando los destinos económicos de los créditos
Destino = customers.groupBy("DESTINOECONOMICO").count()
display(Destino)
```

```
Depto = customers.groupBy("DEPARTAMENTODOMICILIO").count()
display(Depto)
```

La mayoría de créditos se focaliza en compra de vivienda y se solicita en la ciudad de Bogotá.

```
DestinoDept = (customers
              .groupBy('DEPARTAMENTODOMICILIO', 'DESTINOECONOMICO')
              .agg(f.count("DESTINOECONOMICO").alias("count")))
display(DestinoDept)
```

```
#¿Cuál es la distribución por Departamento de acuerdo al destino económico del crédito?

import pyspark.sql.functions as f
from pyspark.sql import Window

w = Window.partitionBy('DEPARTAMENTODOMICILIO')
distr = DestinoDept.withColumn('maxcnt', f.max('count').over(w))\
    .where(f.col('count') == f.col('maxcnt'))\
    .drop('maxcnt')

display(distr)
```

```
from pyspark.sql.functions import *

from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer, OneHotEncoderEstimator
indexer = StringIndexer(inputCol="LINEACREDITO", outputCol="LINEACREDITOIndex", handleInvalid="keep")
encoder = OneHotEncoderEstimator(inputCols=["LINEACREDITOIndex"], outputCols=["LINEACREDITOCategorical"])
pipeline = Pipeline(stages=[indexer, encoder])
pipeline = pipeline.fit(customers)
customers = pipeline.transform(customers)

display(customers)
```
