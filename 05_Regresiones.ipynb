# Seaborn: Regresiones

Seaborn viene con funcionalidades para graficar regresiones. Aunque este será un tema posterior, por ahora  cubriremos la función ** lmplot()**.

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## lmplot()


** lmplot ** permite visualizar modelos lineales, pero también permite dividir los gráficos en función de las características, así como también colorear el tono en función de las características.


```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

### Marcadores

Es posible definir diferentes tipos de marcadores, de acuerdo a los marcadores que soporta matplotlib:

```python
# http://matplotlib.org/api/markers_api.html
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',
           markers=['o','v']) # v en este caso es triángulos
```

Si se quiere modificar otros parámetros de matplotlib más específicos como el tamaño del marcador, es necesario pasar esta información como un parámetro adicional que es un diccionario. Así:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',
           markers=['o','v'],scatter_kws={'s':100}) # scatter_kws, un conjunto de parámetros adicionales de matplotlib, 
# en este caso size, que no se encuentra dentro los de la función
```

**Explicación: **Los argumentos kwargs de lmplot pasan a **regplot** que es una forma más general de lmplot(). regplot tiene un parámetro scatter_kws que se pasa a plt.scatter. Entonces si se quiere establecer el parámetro 's' en ese diccionario, que corresponde al tamaño del marcador, es necesario modificar este argumento en el diccionario. En otras palabras, terminas pasando un diccionario con los argumentos base de matplotlib, en este caso, 's' para el tamaño de un diagrama de dispersión. Si quiere conocer los detalles se recomienda revisar las documentación: http://matplotlib.org/api/markers_api.html


## Cuadrículas

Podemos separar aún más las representaciones de los modelos lineales añadiendo una cuadrícula para analizar variables específicas. Simplemente hay que indicar los argumentos col o row: (El lmplot también permite usar cuadrículas)

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm')
```

## Aspecto y Tamaño

Las figuras de Seaborn pueden ajustarse en tamaño y relación de aspecto con los parámetros **size** y **aspect**:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm',
          aspect=0.6,size=8) # aspect < 1 gráfica mas alta > 1 mas ancha
```
